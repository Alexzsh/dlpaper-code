## 3.2 用SVM建立线性分类器

### 3.2.1 准备工作

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('data_multivar.txt',header=None)
X=np.array(df.iloc[:,:-1])
y=np.array(df.iloc[:,-1])
```

```python
class_0=np.array(df[df[2]==0].iloc[:,:-1])
class_1=np.array(df[df[2]==1].iloc[:,:-1])
```

```python
plt.scatter(class_0[:,0],class_0[:,1],marker='s')
plt.scatter(class_1[:,0],class_1[:,1],marker='o')
```

### 3.2.2 详细步骤

```python
from sklearn.cross_validation import train_test_split
from sklearn.svm import SVC
#切分数据集
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=5)
#定义分类器
params_linear={'kernel':'linear'}
classify_linear=SVC(**params_linear)
classify_linear.fit(X_train,y_train)
```

```python
#画出分类网格边界
def plot_classifier(classifier,X,y,title,annotate=False):
    #划定边界
    x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
    y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
    #创建网格数据
    x_values,y_values=np.meshgrid(np.arange(x_min,x_max,0.01),np.arange(y_min,y_max,0.01))
    mesh_output=classifier.predict(np.c_[x_values.ravel(),y_values.ravel()])
    mesh_output=mesh_output.reshape(x_values.shape)
    
    plt.title(title)
    plt.pcolormesh(x_values,y_values,mesh_output,cmap=plt.cm.RdBu_r)
    
    plt.scatter(X[:,0],X[:,1],c=y,marker='o', edgecolors='black', linewidth=1, cmap=plt.cm.Paired_r)
    
    #给出具体坐标提示
    if annotate:
        for x, y in zip(X[:, 0], X[:, 1]):
            # Full documentation of the function available here:
            # http://matplotlib.org/api/text_api.html#matplotlib.text.Annotation
            plt.annotate(
                '(' + str(round(x, 1)) + ',' + str(round(y, 1)) + ')',
                xy = (x, y), xytext = (-15, 15),
                textcoords = 'offset points',
                horizontalalignment = 'right',
                verticalalignment = 'bottom',
                bbox = dict(boxstyle = 'round,pad=0.6', fc = 'white', alpha = 0.8),
                arrowprops = dict(arrowstyle = '-', connectionstyle = 'arc3,rad=0'))
```

```python
plot_classifier(classify_linear,X_train,y_train,'SVC classify',annotate=False)
```

```python
y_pred=classify_linear.predict(X_test)
plot_classifier(classify_linear,X_test,y_test,'Test')
```

```python
# 查看性能报告
from sklearn.metrics import classification_report
print(classification_report(y_train,classify_linear.predict(X_train)))
print(classification_report(y_test,classify_linear.predict(X_test)))
```

### 3.3 SVM建立非线性分类器
使用多项式核或者RBF高斯核

```python
#degree 越大 则多项式次数越高 拟合计算越大，时间越长
params_poly={'kernel':'poly','degree':3}
classify_poly=SVC(**params_poly)
classify_poly.fit(X_train,y_train)
plot_classifier(classify_poly,X_train,y_train,'Poly')
```

```python
# 与线性分类相比提升了很多
print(classification_report(y_train,classify_poly.predict(X_train)))
```

```python
#更换rbf核
params_rbf={'kernel':'rbf'}
classify_rbf=SVC(**params_rbf)
classify_rbf.fit(X_train,y_train)
plot_classifier(classify_rbf,X_train,y_train,'RBF')
```

```python
# 与Poly核相比提升了很多
print(classification_report(y_train,classify_rbf.predict(X_train)))
```

## 3.4 解决类型数量不平衡问题
到目前为止，我们处理的问题都是所有类型的数据点数量比较接近的情况，但是在真实世界 中，我们不可能总能获取到这么均衡的数据集。有时，某一个类型的数据点数量可能比其他类型 多很多，在这种条件下训练的分类器就会有偏差。边界线不会反映数据的真实特性，因为两种类 型的数据点数量差别太大。因此，需要慎重考虑这种差异性，并想办法调和，才能保证分类器是 不偏不倚的。

```python
df=pd.read_csv('data_multivar_imbalance.txt',header=None)
class_0=np.array(df[df[2]==1].iloc[:,:-1])
class_1=np.array(df[df[2]==0].iloc[:,:-1])
X=np.array(df.iloc[:,:-1])
y=np.array(df.iloc[:,-1]
plt.scatter(class_0[:,0],class_0[:,1],marker='o')
plt.scatter(class_1[:,0],class_1[:,1],marker='o')
```

```python
# 分类器无法区分某一类 导致某一类的准确率为0
classify_linear.fit(X,y)
plot_classifier(classify_linear,X,y,'linear for unbalance')
```

```python
print(classification_report(y,classify_linear.predict(X)))
```

```python
# 对线性分类器进行调整 将class_weight 改成auto 
#使其可以统计不同类型数据的数量 根据此调整对应权重，减小数据分布不均对分类器性能的影响
params_linear['class_weight']='balanced'#官网api已经将auto改为balanced
classify_linear=SVC(**params_linear)
classify_linear.fit(X,y)
plot_classifier(classify_linear,X,y,'auto weight for unbalance')
```

## 3.6 寻找超参数
使用grid_search对参数进行遍历

```python
from sklearn.grid_search import GridSearchCV
parameter_grid=[
    {'kernel':['linear'],'C':[1,10,50,600]},
    {'kernel':['poly'],'degree':[2,3]},
    {'kernel':['rbf'],'gamma':[1e-2,1e-3],'C':[1,10,50,600]}
]
metrics=['precision','recall_weighted']
for metric in metrics:
    print('for every metric' , metric)
    classify=GridSearchCV(SVC(C=1),parameter_grid,cv=5,scoring=metric)
    classify.fit(X_train,y_train)
```

```python
print('Score acoress the parameter grid')
for params,avg_score,_ in classify.grid_scores_:
    print(params,'-->',round(avg_score,3))
```

```python
y_pred=classify.predict(X_test)
print('X_test classification report\n',classification_report(y_test,y_pred))
```


## 3.8 估算交通流量
### 3.8.1 准备工作
我 们 将 要 使 用 的 数 据 集 可 以 在 https://archive.ics.uci.edu/ml/datasets/Dodgers+Loop+Sensor 下 载。这个数据集统计了洛杉矶道奇棒球队(Los Angeles Dodgers)进行主场比赛期间，体育场周 边马路通过的车辆数量，存放在traffic_data.txt文件中。每一行都包含用逗号分隔的字符串格式， 如下所示:  
星期  
时间  
 对手球队  
 棒球比赛是否正在继续  
 通行汽车的数量

```python
traffic_df=pd.read_csv('traffic_data.txt',header=None)
```

```python
X=np.array(traffic_df)
label_encoder=[]
X_encoded=np.empty(X.shape)
for i , item in enumerate(X[0]):
    if str(item).isdigit():
        X_encoded[:,i]=traffic_df.iloc[:,i]
    else:
        label_encoder.append(LabelEncoder())
        X_encoded[:,i]=label_encoder[i].fit_transform(X[:,i])
    print(i,item,X_encoded[:,i])
X=X_encoded[:,:-1].astype(int)
y=X_encoded[:,-1].astype(int)
```

```python
from sklearn.svm import SVR
params_rbf={'kernel':'rbf','C':10.0,'epsilon':0.2}
regressor=SVR(**params_rbf)
regressor.fit(X,y)
```

```python
import sklearn.metrics as sm
print('mae is ',round(sm.mean_absolute_error(y,regressor.predict(X)),2))
```
